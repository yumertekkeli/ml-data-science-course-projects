```python
df.ml = pd.get_dummies(df,drop_first = True)
```

```python
df.ml.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.ml.drop('Churn_Yes',axis = 1)
```

```python
y = df.ml['Churn_Yes']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
base_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,plot_confusion_matrix
```

```python
confusion_matrix(y_test,base_pred)
```

```python
print(classification_report(y_test,base_pred))
```

```python
plot_confusion_matrix(model,X_test,y_test);
```

```python
model.feature_importances_
```

```python
feature_imp = pd.DataFrame(index=X.columns,data=model.feature_importances_,columns=['Feature Importance'])
```

```python
feature_imp.sort_values(by = 'Feature Importance',ascending = False).head()
```
